```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['class']=encoder.fit_transform(data['class'])
data['cap-shape']=encoder.fit_transform(data['cap-shape'])
data['cap-surface']=encoder.fit_transform(data['cap-surface'])
data['cap-color']=encoder.fit_transform(data['cap-color'])
data['bruises']=encoder.fit_transform(data['bruises'])
data['odor']=encoder.fit_transform(data['odor'])
data['gill-attachment']=encoder.fit_transform(data['gill-attachment'])
data['gill-spacing']=encoder.fit_transform(data['gill-spacing'])
data['gill-size']=encoder.fit_transform(data['gill-size'])
data['gill-color']=encoder.fit_transform(data['gill-color'])
data['stalk-shape']=encoder.fit_transform(data['stalk-shape'])
data['stalk-root']=encoder.fit_transform(data['stalk-root'])
data['stalk-surface-above-ring']=encoder.fit_transform(data['stalk-surface-above-ring'])
data['stalk-surface-below-ring']=encoder.fit_transform(data['stalk-surface-below-ring'])
data['stalk-color-above-ring']=encoder.fit_transform(data['stalk-color-above-ring'])
data['stalk-color-below-ring']=encoder.fit_transform(data['stalk-color-below-ring'])
data['veil-type']=encoder.fit_transform(data['veil-type'])
data['veil-color']=encoder.fit_transform(data['veil-color'])
data['ring-number']=encoder.fit_transform(data['ring-number'])
data['ring-type']=encoder.fit_transform(data['ring-type'])
data['spore-print-color']=encoder.fit_transform(data['spore-print-color'])
data['population']=encoder.fit_transform(data['population'])
data['habitat']=encoder.fit_transform(data['habitat'])
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

# Naive Bayes Classifier

## i) Fitting Guassian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,y_train)
```

## Predicting the test result

```python
y_pred=gnb.predict(X_test)
```

## Confusion matrix and Accuracy Score

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)
cm
```

```python
acc
```

## ii) Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
bnb=BernoulliNB()
bnb.fit(X_train,y_train)
```

### Predicting the test result

```python
y_pred=bnb.predict(X_test)
```

## Confusion matrix and Accuracy Score

```python
cm1=confusion_matrix(y_test,y_pred)
acc1=accuracy_score(y_test,y_pred)
cm1
```
